```python
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                
        if df[column].nunique() < 30:       
                                                    
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         
from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "Continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
def info_of_cat(col):
    print(f'Unique values in {col} are: {df[col].unique()}')
    print(f'Mode of {col} is: {df[col].mode()[0]}')
    print(f'missing values in {col} is: {df[col].isnull().sum()}')
    
```

```python
def info_of_numerical(col):
    print(f'mean of the {col} is {df[col].mean()}')
    print(f'median of the {col} is {df[col].median()}')
    print(f'mode of the {col} is {df[col].mode([0])}')
    print(f'standard deviation of the {col} is {df[col].std()}')
    print(f'missing values of the {col} is {df[col].isnull().sum()}')
```
